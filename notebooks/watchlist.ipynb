The purpose of this notebook is to show how to run a query from the API that involves a watchlist.

```python
import os, sys, time, json
from lasair import LasairError, lasair_client as lasair

API_TOKEN = os.getenv('LASAIR_LSST_TOKEN')
if API_TOKEN is None:
    print("No Token found. Check Spelling. Note that if you have just added your token to your environment variables, you may need to restart your terminal so your shell settings are reloaded. ")
```

```python
endpoint = 'https://api.lasair.lsst.ac.uk/api'
L = lasair(API_TOKEN, endpoint=endpoint)
```

### Make an SQL query
We are going to make an SQL query that includes a watchlist. 
Watchlist 6 is the Gaia DR3 white dwarfs.
Click the link to find out about the watchlist.

```python
url = 'https://lasair.lsst.ac.uk/'
wl_id = 31
# Globular cluster watchlist of 157 sources
print(f'See here for info: {url}/watchlists/%d/' % wl_id)
```

```python
selected = """
objects.diaObjectId, watchlist_hits.name, watchlist_hits.arcsec
"""

# Join objects and watchlist 
tables = 'objects, watchlist:%d' % wl_id

conditions = ''

L = lasair(API_TOKEN, endpoint = endpoint)
try:
    results = L.query(selected, tables, conditions, limit = 10)
except LasairError as e:
    print(e)
```

```python
for row in results:
    print(f'{url}/object/{row['diaObjectId']} {row['name']} {row['arcsec']}')
```
