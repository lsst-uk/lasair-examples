```python
import os, sys, json
from lasair import LasairError, lasair_client as lasair

API_TOKEN = os.getenv('LASAIR_LSST_TOKEN')
if API_TOKEN is None:
    print("No Token found. Check Spelling. Note that if you have just added your token to your environment variables, you may need to restart your terminal so your shell settings are reloaded. ")
```

```python
endpoint = "https://api.lasair.lsst.ac.uk/api"
L = lasair(API_TOKEN, endpoint=endpoint)
```

This function simply prints the first lines of the given string.
You can replace it with the normal print if you want to see everything.

```python
def print15(s):
    print('\n'.join(s.split('\n')[:15]))
```

## Sherlock Object

```python
lite=True
objectId = 170028510770888728
result = L.sherlock_object(objectId, lite=lite)
print15(json.dumps(result, indent=2))
```

## Sherlock Position

```python
ra  = 150.778316
dec =   3.343371
result = L.sherlock_position(ra, dec, lite=lite)
print15(json.dumps(result, indent=2))
```
