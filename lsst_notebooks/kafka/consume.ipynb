```python
import sys, time, json
from lasair import lasair_consumer
```

## Normal Consumer

```python
kafka_server = 'lasair-lsst-kafka.lsst.ac.uk:9092'   # prod
topic        = "lasair_1Lightweight"    # prod
```

```python
group_id     = 'test127'
consumer = lasair_consumer(
    kafka_server,    
    group_id,        # any string
    topic            # topic name for your filter
)

n = 0
while n < 10:
    msg = consumer.poll(timeout=5)
    if msg is None:
        print('sleeping 5')
        time.sleep(5)
        continue
    if msg.error():
        print(str(msg.error()))
        break
    jmsg = json.loads(msg.value())
    print(json.dumps(jmsg, indent=2))
    n += 1
print('No more messages available')
```
